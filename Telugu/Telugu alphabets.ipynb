```python
consonants = ['క', 'ఖ', 'గ', 'ఘ', 'ఙ', 'చ', 'ఛ', 'జ', 'ఝ', 'ఞ', 'ట', 'ఠ', 'డ', 'ఢ', 'ణ', 'త', 'థ', 'ద', 'ధ',
              'న', 'ప', 'ఫ', 'బ', 'భ', 'మ', 'య', 'ర', 'ల', 'వ', 'శ', 'ష', 'స', 'హ', 'ళ', 'క్ష', 'ఱ']

vowels = ['అ', 'ఆ', 'ఇ', 'ఈ', 'ఉ', 'ఊ', 'ఋ', 'ౠ', 'ఎ', 'ఏ', 'ఐ', 'ఒ', 'ఓ', 'ఔ', 'అం', 'అః']

matras = ['ా', 'ి', 'ీ', 'ు', 'ూ', 'ృ', 'ౄ', 'ె', 'ే', 'ై', 'ొ', 'ో', 'ౌ', '్']

all_combinations = []

for matra in matras:
    for vowel in vowels:
        combination = vowel + matra
        all_combinations.append(combination)

for matra in matras:
    for consonant in consonants:
        combination = consonant + matra
        all_combinations.append(combination)

print(all_combinations)
# Few of them might not make sense. Should delete them
```

```python
from PIL import Image, ImageDraw, ImageFont
from tqdm import tqdm
```

```python
# Calculate text size and position
for aksharam in tqdm(all_combinations):
    width, height = 100, 100  # Image size
    background_color = 'white'
    font_color = 'black'
    image = Image.new('RGB', (width, height), color=background_color)
    draw = ImageDraw.Draw(image)
    
    # Load a Telugu font
    font_path = '/usr/share/fonts/truetype/fonts-telu-extra/vemana2000.ttf'
    font_size = 60
    font = ImageFont.truetype(font_path, font_size)
    
    text_width, text_height = draw.textsize(aksharam, font=font)
    x = (width - text_width) // 2
    y = (height - text_height) // 2
    
    # Draw the Telugu character on the image
    draw.text((x, y), aksharam, fill=font_color, font=font)
    # Save the image
    image.save('aksharalu/'+aksharam+'.png')
```
